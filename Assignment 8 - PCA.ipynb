```python
# Converting data to numpy array

wine_arr = new_wine.values
wine_arr
```

```python
from sklearn.preprocessing import scale
```

```python
# Normalizing the numerical data 

wine_norm = scale(wine_arr)
wine_norm
```

### Implementing PCA

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=13)
```

```python
wine_pca = pca.fit_transform(wine_norm)
wine_pca
```

```python
pca.components_
```

```python
var = pca.explained_variance_ratio_
var
```

```python
# Cummulative variance

var1 = np.cumsum(np.round(var,4)*100)
var1
```

```python
# Variance plot for PCA components

plt.plot(var1,color='magenta')
```

```python
final_df = pd.concat([wine['Type'],pd.DataFrame(wine_pca[:,0:3],columns=['PC1','PC2','PC3'])],axis=1)
final_df
```

```python
# Visualization of PCAs

fig = plt.figure(figsize=(16,12))
snb.scatterplot(data=final_df)
```

### Checking with other Methods

```python
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import normalize
import warnings
warnings.filterwarnings('ignore')
```

```python
plt.figure(figsize=(10,8))
dendrogram = sch.dendrogram(sch.linkage(wine_norm,'complete'))
```

```python
hierar_clusters = AgglomerativeClustering(n_clusters=3,affinity='euclidean',linkage='ward')
hierar_clusters
```

```python
y = pd.DataFrame(hierar_clusters.fit_predict(wine_norm),columns=['clustersid'])
y['clustersid'].value_counts()
```

```python
# Adding clusters to dataset

wine3 = wine.copy()
```

```python
wine3['clustersid'] = hierar_clusters.labels_
wine3
```

### K-Means

```python
from sklearn.cluster import KMeans
```

```python
wcss = []
for i in range (1,6):
    kmeans = KMeans(n_clusters=i,random_state=2)
    kmeans.fit(wine_norm)
    wcss.append(kmeans.inertia_)
```

```python
# Plotting K values vs WCSS to get Elbow graph for choosing K (no. of clusters)

plt.plot(range(1,6),wcss)
plt.title('Elbow Graph')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

### Building Cluster Algo using K = 3

```python
clusters3 = KMeans(3,random_state=30).fit(wine_norm)
clusters3
```

```python
clusters3.labels_
```

```python
# Assign clusters to the data set

wine4 = wine.copy()
```

```python
wine4['clusters3id'] = clusters3.labels_
wine4
```
